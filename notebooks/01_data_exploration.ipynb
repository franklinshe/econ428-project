```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/bank/bank-full.csv', delimiter=';')
```

```python
# Calculate the proportion of each class
class_proportion = df['y'].value_counts(normalize=True)

# Visualize the class imbalance
plt.figure(figsize=(8, 6))
sns.barplot(x=class_proportion.index, y=class_proportion.values)
plt.title('Proportion of Classes in Target Variable')
plt.ylabel('Proportion')
plt.xlabel('Subscription Outcome')
plt.show()
```

```python
# Example for the 'job' feature
plt.figure(figsize=(12, 8))
sns.countplot(y='job', hue='y', data=df)
plt.title('Count of Subscriptions by Job Title')
plt.xlabel('Count')
plt.ylabel('Job Title')
plt.legend(title='Subscription Outcome')
plt.show()
```

```python
data = df[['age', 'balance', 'duration', 'pdays', 'previous']]

# Set up the matplotlib figure
plt.figure(figsize=(12, 6))

# Create a boxplot for each variable in the DataFrame
for i, column in enumerate(data.columns, 1):
    plt.subplot(1, 5, i)
    sns.boxplot(y=data[column])
    plt.title(column)

plt.tight_layout()
plt.show()
```

```python
# Create a new DataFrame for the log-transformed data
data_log = data.copy()

data_log['age'] = np.log1p(data_log['age'])
data_log['balance'] = np.log1p(data_log['balance'] + np.abs(data_log['balance'].min()) + 1)
data_log['duration'] = np.log1p(data_log['duration'])
data_log['pdays'] = np.log1p(data_log['pdays'] + 1)
data_log['previous'] = np.log1p(data_log['previous'])

# Set up the matplotlib figure
plt.figure(figsize=(12, 6))

# Create a boxplot for each variable in the log-transformed DataFrame
for i, column in enumerate(data_log.columns, 1):
    plt.subplot(1, 5, i)
    sns.boxplot(y=data_log[column])
    plt.title(column)
    plt.ylabel('Log Scale')

plt.tight_layout()
plt.show()
```
